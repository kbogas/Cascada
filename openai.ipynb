```python
import numpy as np
initial = [0.05, 0, 0.03, 0.76, 0.07, 0.09]
transitions = np.array([[0.73, 0.02, 0.04, 0.07, 0.06, 0.08],
                       [0.01, 0.06, 0.07, 0.07, 0.08, 0.71],
                       [0.09, 0.08, 0.09, 0.71, 0.03, 0],
                       [0.05, 0, 0.02, 0.84, 0.08, 0.01],
                       [0.03, 0.8, 0.07, 0, 0.01, 0.09],
                       [0, 0.04, 0.07, 0.03, 0.79, 0.07]])
```

```python
from copy import deepcopy
steps = 5
tr = deepcopy(transitions)
for st in xrange(steps):
    if st == 0:
        tr = deepcopy(transitions)
    else:
        tr = np.multiply(tr, transitions)
```

```python
np.dot(tr, initial).shape
```

```python
trials = 1000
states = [l for l in 'openai']
steps = 5
results = {}
for i in xrange(trials):
    seq = ''
    p = 1
    for step in xrange(steps):
        if step == 0:
            cur_state = np.random.choice(len(states),1, p=initial)[0]
            seq +=states[cur_state]
            p *= initial[cur_state]
        else:
            prev_state = cur_state
            cur_state = np.random.choice(len(states),1, p=transitions[prev_state,:].tolist())[0]
            p *= transitions[prev_state,:].tolist()[cur_state]
            seq +=states[cur_state]
    if seq in results:
        results[seq]['count'] += 1
    else:
        results[seq] = {'count':1, 'prob':p}
    
        
```

```python
import pandas as pd
df = pd.DataFrame(results).T
df.sort_values('prob', ascending=False)
```
